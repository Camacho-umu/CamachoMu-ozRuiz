# Aprendizaje por Refuerzo en Entornos Complejos



## Introducción

En esta segunda parte del proyecto, abordamos el problema de la toma de decisiones secuenciales en entornos complejos modelados como Procesos de Decisión de Markov (MDP). A diferencia del problema del bandido multibrazo, aquí las decisiones del agente afectan no solo a la recompensa inmediata, sino también a los estados futuros.

Para resolver estos entornos sin conocer a priori las dinámicas de transición ni las recompensas exactas, hemos implementado y comparado técnicas de **Aprendizaje por Refuerzo Tabular** y de **Control con Aproximaciones**.
